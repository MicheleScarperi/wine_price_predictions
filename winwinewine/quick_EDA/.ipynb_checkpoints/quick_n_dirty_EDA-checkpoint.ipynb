```python
labels = list(df.columns.values)

for label in labels:
    print(df[label].value_counts())
    
```

```python
data = df[['price','title']]
#print(data)
  
df = pd.DataFrame(data,columns=['price','title'])
print (df.info)

#df.drop_duplicates()
#print('dataframe without duplicates: ',df.info)
# --> there are no duplicates

titles = df['title'][:100]
prices = df['price'][:100]
```

```python
# preprocessing
print('before dropping NaN: ', df.info) # 129971 rows x 14 columns
df.dropna(inplace=True)
print('after dropping NaN values in rows: ', df.info) # down to 22387 rows x 14 columns
```

```python
# dropping some columns for now that need special treatment - if at all useful. 
df.drop(['Unnamed: 0','description','taster_twitter_handle','title'],inplace=True,axis=1)
df.shape
print(labels,len(labels))

#for label in labels:
    #print(type(df[label]))
df.info()
```

```python
#label encoding the data before onehotencoding it (this is required in case of text data)
#print('before: ',df['country'].head())
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['country']= le.fit_transform(df['country']) 
df['designation']= le.fit_transform(df['designation'])
df['province']= le.fit_transform(df['province'])
df['region_1']= le.fit_transform(df['region_1'])
df['region_2']= le.fit_transform(df['region_2'])
df['taster_name']= le.fit_transform(df['taster_name'])
df['variety']= le.fit_transform(df['variety'])
df['winery']= le.fit_transform(df['winery'])

#print('after: ',df['country'].head())
#for label in labels:
    #print(type(df[label].dtype))
    
df.info()
```

```python
# now we have to drop the price column and hotencode the rest of the df. 

df_wout_price = df.drop(['price'],axis=1)

print(df_wout_price.info())

from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder() 
  
data = onehotencoder.fit_transform(df_wout_price).toarray()

print(data)

```

```python
# lets split the data now and start testing what results we can get with the currently preprocessed data. 
X = data
y = df['price']
print(X.shape,y.shape)

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)

#print(df['price'].head())
```

```python
# i am trying a linear regression model first just for testing overall. 

from sklearn.linear_model import LinearRegression

lr = LinearRegression().fit(X_train,y_train)

print('training score: ',lr.score(X_train,y_train),'test score: ',lr.score(X_test,y_test))
```
